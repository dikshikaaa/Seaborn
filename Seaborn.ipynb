## Seaborn
*   Data Visualisation Library






```python
import seaborn as sns
import matplotlib as plt
import numpy as np
import pandas as pd
```

Seaborn has some built-in data sets

```python
#total bill vs tip dataset
tips = sns.load_dataset('tips')
# tips will be loaded as pandas table
tips.head()
```

```python
tips.tail()
```

```python
sns.relplot(data = tips, x = 'total_bill', y = 'tip', col = 'time', hue = 'smoker', style = 'day', size = 'size')
```

```python
#visualize the data
sns.relplot(data = tips, x = 'total_bill', y = 'tip', col = 'time', hue = 'smoker', style = 'smoker', size = 'size')
```

```python
#setting theme for the data

sns.set_theme()

sns.relplot(data = tips, x = 'total_bill', y = 'tip', col = 'time', hue = 'smoker', style = 'smoker', size = 'size')
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris.tail()
```

```python
sns.scatterplot(data = iris, x = 'sepal_length', y = 'petal_length', hue = 'species')
```

```python
sns.scatterplot(data = iris, x = 'sepal_length', y = 'petal_width', hue = 'species')
```

```python
#loading the titanic data set
titanic  = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
sns.scatterplot(data = titanic, x = 'age', y = 'class', hue = 'alive', style = 'sex')
```

```python
sns.countplot(data = titanic, x = 'class')
```

```python
sns.countplot(data = titanic, x = 'alive')
```

```python
sns.barplot(x='sex', y='survived', hue = 'class', data = titanic)
```

```python
# house price dataset
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
boston_df = pd.DataFrame(boston.data, columns = boston.feature_names)
boston_df['PRICE'] = boston.target
```

```python
boston_df.head()
```

Distribution Plot

```python
sns.displot(boston_df['PRICE'])
```

```python
sns.distplot(boston_df['PRICE'])
```

```python
sns.countplot(data = boston_df, x ='PRICE')
```
